```python
# Importing the packages
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
import re
import matplotlib.pyplot as plt
import seaborn as sns

import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
nltk.download('stopwords')

from langdetect import detect
```

```python
def clean_lyrics(raw_lyrics:str):
  """
  Removes unrelated characters from the texts.

  Parameters:
  - raw_lyrics (str): The raw text that wanted to be removed unrelated characters.

  Returns:
  string: representing clean version of the raw text.
  """
    # Regex pattern to match section markers
    pattern = r'\[.+]'

    # Replace section markers with an empty string
    cleaned_lyrics = re.sub(pattern, '', raw_lyrics)

    # Remove leading and trailing whitespace and newlines
    cleaned_lyrics = cleaned_lyrics.strip()

    return cleaned_lyrics


def text_to_embeds(text:str, model_sequence_length:int, overlap:int):
  """
  Converts texts into chunks regarding to model sequence length. Then gets embedding representations of them by using average pooling.

  Parameters:
  - text (str): The texts that wanted to be represented as embedding.
  - model_sequence_length (int): The maximum sequence length of the model used.
  - overlap (int): The number of words at the end of one text chunk that will be included at the beginning of the next chunk when splitting the input text into segments for embedding. This parameter ensures continuity and contextual connection between consecutive chunks, especially important for maintaining semantic integrity in natural language processing tasks.
  
  Returns:
  numpy array: representing texts as embeddings.
  """
  splitted_text = text.split()
  chunks = []
  if len(splitted_text) > model_sequence_length:      # If text sequence length is longer than model sequence length, then the text is splitted by chunks
    for i in range(0, len(splitted_text), model_sequence_length-overlap):
      chunk = splitted_text[i:i+model_sequence_length]
      chunks.append(chunk)

      chunk_texts = [" ".join(c) for c in chunks]

      embeddings = [model.encode(chunk_text, convert_to_tensor=True) for chunk_text in chunk_texts]  # Obtaining embeddings of each chunk
      embeddings = np.mean(np.stack(embeddings), axis=0)    # Applying average pooling to all chunks

  else:
    embeddings = model.encode(text, convert_to_tensor=True).numpy()   # If the text sequence length is shorter than model sequence length, then the text is converted to embedding directly

  return embeddings

def outlier_cutoff(dataframe:pd.DataFrame, col:str, low_quantile:float, high_quantile:float):
  """
  Filters the data based on the quartiles of given column.

  Parameters:
  - dataframe (pd.DataFrame): The dataframe that includes column which wanted to be filtered.
  - col (str): The column that wanted to be filtered based on the quartiles.
  - low_quantile (float): Low limit of the quartile of given column.
  - high_quantile (float): Up limit of the quartile of given column.

  Returns:
  pandas dataframe: representing remaining values between the specified quartiles.
  """
  return dataframe[(dataframe[col]<dataframe[col].quantile(high_quantile)) & (dataframe[col]>dataframe[col].quantile(low_quantile))]
```

```python
# Loading Hot 100 Charts data
songs_path = "Billboard_Lists_1960-01-01_2024-02-23.csv"
song_df = pd.read_csv(songs_path)
```

```python
# Removing the whitespaces at the end and start of the values in all string type columns. 
song_df = song_df.applymap(lambda x: x.strip() if type(x) != int else x)
```

```python
# Getting year info from the 'Week' column.
song_df["Year"] = pd.to_datetime(song_df["Week"]).dt.year
```

```python
# Filtering the data since 1990 because of the computing issues.
song_df2 = song_df.loc[song_df["Year"]>=1990].reset_index(drop=True)
```

```python
# Dropping duplicates based on artists and songs
song_df3 = song_df2[["Artist_Name","Song"]].drop_duplicates().reset_index(drop=True)
```

```python
# Loading the lyrics data. This works with TPU source in Colab
file_path = "Billboard_Lyrics_1960-01-01_2024-02-23_Part9.csv"

data = pd.read_csv(file_path)
```

```python
# Removing the whitespaces at the end and start of the values in all string type columns. 
data = data.applymap(lambda x: x.strip() if type(x) not in [float, int] else x)
```

```python
# Selecting the lyrics data based hot 100 charts data.
df = data[(data["Artist_Name"].isin(song_df3["Artist_Name"])) & (data["Song"].isin(song_df3["Song"]))].reset_index(drop=True)
```

```python
# Detecting the language of the lyrics.
from langdetect import detect

df["Language"] = df["Lyrics"].apply(lambda text: detect(text) if pd.notna(text) else np.nan)
```

```python
# Getting all languages in the data
lang_list = df["Language"].unique().tolist()
if np.nan in lang_list:
  lang_list.remove(np.nan)
else:
  pass
```

```python
# Converting language name into long version. For example: 'en':'english'
from iso639 import Lang

lang_names = [Lang(la).name.lower() if len(la.split("-")) <= 1 else Lang(la.split("-")[0]).name.lower() for la in lang_list]
```

```python
lang_map = dict(zip(lang_list, lang_names))
lang_map
```

```python
# Plotting the distribution of the language in dataset before translation.
pie_lang_label, pie_lang_count = df['Language'].value_counts().index.tolist(), df['Language'].value_counts().values.tolist()

plt.figure(figsize=(20,10))


explode = (np.ones(len(pie_lang_label)) * 0.3).tolist()
plt.pie(pie_lang_count, labels=None,autopct='%.0f%%', explode=explode, pctdistance=1.1, labeldistance=1.5)

plt.axis('equal')
plt.legend(labels = pie_lang_label, loc="upper left")
plt.show()
```

### We can see that the 96% of songs' lyrics are english above. We need to translate the rest of lyrics into English.

```python
# Splitting the languages as 'en' and 'Other'
p_df = df["Language"].value_counts().reset_index()
p_df.rename(columns={"index":"Language", "Language":"Count"}, inplace=True)

p_df.loc[p_df["Language"]!="en", "Language"] = "Other"

p_df = p_df.groupby("Language", as_index=False)["Count"].sum()

p_df
```

```python
# Detecting the languages that not supporting by GoogleTranslator
from deep_translator import GoogleTranslator

supported_langs = GoogleTranslator().get_supported_languages()

unsupported_langs = [la for la in lang_names if la not in supported_langs]

unsupported_langs
```

```python
# Unsupported Language Rate

len(df[df["Language"].map(lang_map).isin(unsupported_langs)]) / len(df)
```

```python
# Translating the lyrics into English by using GoogleTranslator.
df["Translated_Lyrics"] = np.nan

for i in range(len(df)):
  try:
    print(i)
    if df["Language"][i] == "en":
      df["Translated_Lyrics"][i] = df["Lyrics"][i]
    else:
      translated = GoogleTranslator(source='auto', target='en').translate(df["Lyrics"][i])
      df["Translated_Lyrics"][i] = translated
  except:
    continue
```

```python
# Saving and Loading the translated lyrics.
#df.to_csv("Translated_Lyrics.csv")
df = pd.read_csv("Translated_Lyrics.csv")
df = df.drop("Unnamed: 0", axis=1)
```

```python
# Checking the language again whether english lyrics increase.
from langdetect import detect

df["Translated_Language"] = df["Translated_Lyrics"].apply(lambda text: detect(text) if pd.notna(text) else np.nan)
```

```python
# Plotting the distribution of the language in dataset after translation.
pie_lang_label, pie_lang_count = df['Translated_Language'].value_counts().index.tolist(), df['Translated_Language'].value_counts().values.tolist()

plt.figure(figsize=(20,10))


explode = (np.ones(len(pie_lang_label)) * 0.3).tolist()
plt.pie(pie_lang_count, labels=None,autopct='%.0f%%', explode=explode, pctdistance=1.1, labeldistance=1.5)

plt.axis('equal')
plt.legend(labels = pie_lang_label, loc="upper left")
plt.show()
```

### After translating the non-english lyrics into english, we obtained almost 100% of english songs' lyrics.

```python
# Splitting the languages as 'en' and 'Other' in after translation version.
a_p_df = df["Translated_Language"].value_counts().reset_index()
a_p_df.columns = ["Language","Count"]
a_p_df.loc[a_p_df["Language"]!="en", "Language"] = "Other"
a_p_df = a_p_df.groupby("Language", as_index=False)["Count"].sum()
a_p_df
```

```python
# Merging before and after translation datasets.
merge_p_df = pd.merge(p_df, a_p_df, on = "Language", how="inner")
```

```python
# Comparing distributions of before and after translation versions of the lyrics. 
import plotly.graph_objects as go
from plotly.subplots import make_subplots

labels= merge_p_df["Language"].tolist()
# Create subplots: use 'domain' type for Pie subplot
fig = make_subplots(rows=1, cols=2, specs=[[{'type':'domain'}, {'type':'domain'}]])
fig.add_trace(go.Pie(labels=labels, values=merge_p_df["Count_x"].values, name="Before"),
              1, 1)
fig.add_trace(go.Pie(labels=labels, values=merge_p_df["Count_y"].values, name="After"),
              1, 2)

# Use `hole` to create a donut-like pie chart
fig.update_traces(hole=.4, hoverinfo="label+percent+name")

fig.update_layout(
    title_text="Lyrics Translation",
    # Add annotations in the center of the donut pies.
    annotations=[dict(text='Before', x=0.20, y=0.5, font_size=20, showarrow=False),
                 dict(text='After', x=0.80, y=0.5, font_size=20, showarrow=False)])
fig.show()
```

### Above pie chart we can observe the change of english lyrics distribution after translating process.

```python
# Data Cleaning by removing unrelated characters from the lyrics.

tokenizer = RegexpTokenizer(r'[a-zA-Z]+')

df["Translated_Lyrics_2"] = df["Translated_Lyrics"].apply(lambda text: "\n".join(text.split("\n")[1:]) if pd.notna(text) else text)     # Removing song title from the lyrics
df["Tokenized_Text"] = df["Translated_Lyrics_2"].apply(lambda text: clean_lyrics(text) if pd.notna(text) else text)
df["Tokenized_Text"] = df["Tokenized_Text"].apply(lambda x: x.replace("*vocalizations*", "").strip() if pd.notna(x) else x)
df["Tokenized_Text"] = df["Tokenized_Text"].str.rstrip("Embed")
df["Tokenized_Text"] = df["Tokenized_Text"].apply(lambda text: re.sub(r'Contributors|Lyrics|Contributor', '', text) if pd.notna(text) else text)
df["Tokenized_Text"] = df["Tokenized_Text"].apply(lambda text: " ".join(tokenizer.tokenize(text.lower())) if pd.notna(text) else text)   # Regex Tokenization
```

## ENGLISH SONGS

```python
# Selecting english songs to analyze
en_df = df[df["Translated_Language"]=="en"].reset_index(drop=True)
```

```python
# Comparing the number of english songs with the all songs
len(en_df), len(df)
```

```python
# Detecting the lyrics length.
en_df["Sentence_Length"] = en_df["Tokenized_Text"].apply(lambda text: len(text.split()) if pd.notna(text) else np.nan)
```

```python
# Checking the quartiles of the lyrics length.
en_df["Sentence_Length"].describe([0.05,0.1,0.25,0.5,0.75,0.85,0.9,0.95,0.99])
```

### Above distribution data, we can observe that some of the rows in the data have 0 sentence length and some of them also have so many words. We need to remove them as outliers from the data to get more objective analysis.

```python
# Applying outlier_cutoff function that is defined above to filter based up and low quartile of the lyrics length.
final_df = outlier_cutoff(en_df, "Sentence_Length", 0.05, 0.95)
final_df = final_df.reset_index(drop=True)
```

```python
# After Outlier Threshold

pie_data = [len(final_df), len(en_df) - len(final_df)]
pie_label = ["Final Data", "Deleted Part"]

plt.pie(pie_data, labels=pie_label, autopct='%.0f%%')
plt.show();
```

### After eliminating the data according their sentence length, we had to remove 10% of data to continue our analysis.

```python
# Loading pretrained sentence transformer embedding model. 'all-mpnet-base-v2' is used for Sentiment Analysis, while 'all-MiniLM-L6-v2' is used for Topic Modeling.
from sentence_transformers import SentenceTransformer

transformers_name = "all-mpnet-base-v2"

# Load a pre-trained Sentence Transformer model
model = SentenceTransformer(transformers_name)
max_seq_length = model.max_seq_length
```

```python
# Lyrics length distributions after outlier cutoff process.
final_df["Sentence_Length"].describe()
```

# Now, we have more normal sentence lengths.

```python
# Getting embeddings of all lyrics.
from tqdm import tqdm
import time

start_time = time.time()
overlap=10
model_embeds_dimension = model.get_sentence_embedding_dimension()
embeds_arr = np.empty(shape=(len(final_df), model_embeds_dimension))

for i in tqdm(range(len(final_df["Tokenized_Text"])), desc="Embedding Texts"):
  text_embedding = text_to_embeds(final_df["Tokenized_Text"][i], max_seq_length, overlap)

  embeds_arr[i] = text_embedding

end_time = time.time()

print(f"Process Time : {(end_time-start_time)/60} minutes")
```

```python
# Saving Embeddings
np.save('Embeds_Lyrics_20240309.npy', embeds_arr)
```

```python
# Saving the final version of the data.
final_df.to_csv("Final_Data_20240309.csv")
```
